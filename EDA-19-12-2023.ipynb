$one-hot-encoder$

```python
-one hot encoder means at a time only one will be ON(1/TRUE) others are OFF(0/FALSE)
-suppose case_status has two unique labels
-certified and denied
-in one hot encoder it will create two new columns
-like below
```

|case_status|certified|denied|
|-----------|---------|------|
|certified|1|0|
|denied|0|1|

```python
-one hot encoder new columns are orthogonal each other
-orthogonal means 90 degrees phase shift
-90 degrees phase shift means perpendicular each other
-perpendicular means independent each other
-independent means no relation
-it is very important property  that is to avoid realtion between variables
```

$drawback$

```python
-assume that you have 100 unique labels for a column
-so it will create 100 new columns
-to process 100 columns we require more memory,more time,more hardware
-this is called curse of dimensionality
```

$z-score$
<img class="mimg" style="color: rgb(102, 102, 102);" height="180" width="303" src="https://th.bing.com/th/id/OIP.mmO8G4jCsp8ifwBSOqkT7wHaED?w=303&amp;h=180&amp;c=7&amp;r=0&amp;o=5&amp;dpr=1.3&amp;pid=1.7" alt="Image result for standardization formula" data-thhnrepbd="1" data-bm="109">

```python
# step-1:calculate min value of p_wage=minn
# step-2:maxx of p_wage
# step-3:dr=maxx/minn
# step-4:nr=p_wage/minn
# step-5:nr/dr
```

```python
mean=visa_df["prevailing_wage"].mean()
std=visa_df["prevailing_wage"].std()
z=visa_df["prevailing_wage"]-mean
visa_df["prevailing_wage_zscore"]=z/std
```

```python
visa_df[["prevailing_wage","prevailing_wage_zscore"]]
```

```python
plt.subplot(1,2,1)
plt.boxplot(visa_df["prevailing_wage"])
plt.subplot(1,2,2)
plt.boxplot(visa_df["prevailing_wage_zscore"])
plt.show()
```

```python
q1=np.percentile(visa_df["prevailing_wage_zscore"],25)
q2=np.percentile(visa_df["preaviling_wage_zscore"],50)
q3=np.percentile(visa_df["prevailing_wage_zscore"],75)
iqr=q3-q1
ub=q3+1.5*iqr
lb=q1-1.5*iqr
con1=visa_df["prevailing_wage_zscore"]>ub
con2=visa_df["prevailing_wage_zscore"]<lb
con=con1|con2
visa_df["prevailing_wage_zscore"][con]
```
