```python
#m-1: using matplotlib
===================================reading the data==========================================================
file_path="C:\\Users\\shrav\\Downloads\\Visadataset - Visadataset.csv"
visa_df=pd.read_csv(file_path)
===================================create a frequency table==================================================
visa_df["continent"].value_counts()
continents=visa_df["continent"].value_counts().keys()
count=visa_df["continent"].value_counts().values
contint_data=pd.DataFrame(zip(continents,count),columns=["continents","count"])
contint_data
===================================plot======================================================================
plt.figure(figsize=(10,5))
#10=x-axis
#5=y-axis
plt.bar("continents","count",data=contint_data)
plt.title("Bar Chart")
plt.xlabel("continents")
plt.ylabel("count")
plt.savefig("continents_bar.jpg")
plt.show()
```

$pie-chart$

```python
keys=visa_df["continent"].value_counts(normalize=True).keys()
values=visa_df["continent"].value_counts(normalize=True).values
pd.DataFrame(zip(keys,values),columns=["continent","Relative frequency"])
```

```python
plt.pie(x=values,labels=keys,autopct="%0.2f%%",startangle=360,radius=3,explode=[0.1,2,3,4,5,6])
plt.show()
```
